```python
# List of PMIDs you want to process
pmid_list = all_pmid_unique  # <-- replace with your actual list

pdf_dir = "/Users/yiquntchen/Dropbox/causaljudge-data/pdf-data"
output_dir = "/Users/yiquntchen/Dropbox/causaljudge-data/markdown-data"
os.makedirs(output_dir, exist_ok=True)

for pmid in tqdm.tqdm(pmid_list, desc="Processing PDFs"):
    try:
        output_path = os.path.join(output_dir, f"{pmid}.json")

        # Skip if JSON already exists
        if os.path.exists(output_path):
            tqdm.tqdm.write(f"Skipping PMID {pmid}: JSON already exists.")
            continue

        # 1. Prepare the form data
        pdf_path = os.path.join(pdf_dir, f"{pmid}.pdf")

        if not os.path.exists(pdf_path):
            tqdm.tqdm.write(f"Warning: PDF not found for PMID {pmid}, skipping.")
            continue

        with open(pdf_path, "rb") as f:
            form_data = {
                'file': (f'{pmid}.pdf', f, 'application/pdf'),
                'langs': (None, "English"),
                "force_ocr": (None, True),
                "paginate": (None, False),
                'output_format': (None, 'markdown'),
                "use_llm": (None, False),
                "strip_existing_ocr": (None, True),
                "disable_image_extraction": (None, False)
            }

            # 2. Send the POST request
            response = requests.post(url, files=form_data, headers=headers)
            data = response.json()

        # 3. Get the check URL and start polling
        check_url = data["request_check_url"]
        max_polls = 300
        success = False

        for _ in range(max_polls):
            time.sleep(2)
            poll_response = requests.get(check_url, headers=headers)
            poll_data = poll_response.json()

            if poll_data["status"] == "complete":
                success = True
                break

        # 4. Save the result if successful
        if success:
            with open(output_path, "w") as f_out:
                json.dump(poll_data, f_out)
            tqdm.tqdm.write(f"Saved PMID {pmid} result.")
        else:
            tqdm.tqdm.write(f"Warning: PMID {pmid} did not complete in time.")

    except Exception as e:
        tqdm.tqdm.write(f"Error processing PMID {pmid}: {e}")
```

```python
# Directories
input_dir = "/Users/yiquntchen/Dropbox/causaljudge-data/markdown-data/"
output_file = os.path.join("/Users/yiquntchen/Dropbox/causaljudge-data/PMID_all_text.jsonl")

# Find all JSON files in the input directory
json_files = [f for f in os.listdir(input_dir) if f.endswith(".json")]

with open(output_file, "w") as outfile:
    for json_file in tqdm.tqdm(json_files, desc="Processing JSONs"):
        pmid = json_file.replace(".json", "")
        json_path = os.path.join(input_dir, json_file)

        try:
            with open(json_path, "r") as f:
                data = json.load(f)

            markdown_text = data.get("markdown")
            if markdown_text is not None:
                entry = {"pmid": pmid, "text": markdown_text}
                outfile.write(json.dumps(entry) + "\n")
            else:
                tqdm.write(f"Warning: No 'markdown' found in {json_file}, skipping.")
        
        except Exception as e:
            tqdm.write(f"Error processing {json_file}: {e}")
```
